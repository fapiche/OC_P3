```python
#importation des librairies utiles
import pandas as pd
```

```python
#Fonction qui récupère la population d'un pays donné
def recup_pop(pays) :
    pop = data_population[data_population[zone] == pays][valeur].values
    pop = pop * 1000
    return pop
```

```python
#Fonction qui passe les valeurs/jour en valeurs/an
def convertir_kcal_annee(data) :
    data.loc[data[unite] == 'Kcal/personne/jour', valeur] *=  365
    data.loc[data[unite] == 'Kcal/personne/jour', unite] = 'Kcal/personne'
    return data
```

```python
#fonction passe les valeurs/personne en valeurs pour toute la population
def convertir_pop(data) :
    data_annee = convertir_kcal_annee(data)
    for (z, p, u) in zip(data_annee[zone], data_annee[produit], data_annee[unite]) :
        mask1 = data_annee[zone] == z
        mask2 = data_annee[produit] == p
        mask3 = data_annee[unite] == u
        pop = recup_pop(z)
        if u == 'Kcal/personne' :
            data_annee.loc[ mask3 & mask1 & mask2, valeur] *= pop
        elif u == 'Milliers de tonnes' :
            data_annee.loc[ mask3 & mask1 & mask2, valeur] *= 1000000 
    data_annee.loc[data_annee[unite] == 'Kcal/personne', unite] = 'Kcal'
    data_annee.loc[data_annee[unite] == 'Milliers de tonnes', unite] = 'kg'
    return data_annee
        
```

```python
#Récupération des données utiles
data_population = pd.read_csv('fr_population.csv')
data_animaux = pd.read_csv('fr_animaux.csv')
data_vegetaux = pd.read_csv('fr_vegetaux.csv')

#On gère nos noms de colonnes via des variables pour faciliter la maintenance du code
code_zone = "Code zone"
zone = 'Zone'
valeur = 'Valeur'
element = "Élément"
code_domaine = 'Code Domaine'
domaine = 'Domaine'
code_element = 'Code Élément'
code_produit = 'Code Produit'
produit = 'Produit'
code_annee = 'Code année'
annee = 'Année'
symbole = 'Symbole'
description_symbole = 'Description du Symbole'
unite = 'Unité'

data_population = data_population[data_population[code_zone] != 351]
data_animaux = data_animaux[data_animaux[code_zone] != 351]
data_vegetaux = data_vegetaux[data_vegetaux[code_zone] != 351]
```

```python
#Suppression des doublons, des colonnes et des lignes inutiles de nos dataframe
data_population = data_population[data_population[code_zone] != 351]

data_animaux = data_animaux[data_animaux[code_zone] != 351]
data_animaux = data_animaux[data_animaux[element].str.contains('Kcal/personne/jour|Nourriture')]
data_animaux = data_animaux.drop([code_zone, code_element, element, code_domaine, domaine, code_produit, code_annee, annee, symbole, description_symbole], axis = 1)
data_vegetaux = data_vegetaux[data_vegetaux[code_zone] != 351]
data_vegetaux = data_vegetaux[data_vegetaux[element].str.contains('Kcal/personne/jour|Nourriture')]
data_vegetaux = data_vegetaux.drop([code_zone, code_element, element, code_domaine, domaine, code_produit, code_annee, annee, symbole, description_symbole], axis = 1)
```

```python
#On réunit nos jeux de données
data_dispo = data_animaux.append(data_vegetaux)
data_dispo.head(20)
```

```python
data_dispo = convertir_pop(data_dispo)
data_dispo.head(20)
```

```python
data_zone = data_dispo.groupby([zone, unite]).sum(valeur)
data_zone.head()
```
